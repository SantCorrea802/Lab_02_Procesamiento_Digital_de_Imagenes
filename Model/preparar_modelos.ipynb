```python
hog = cv2.HOGDescriptor((128,128), (16,16), (8,8), (8,8), 9) # configuramos el hog
                        # ventana, bloque, stride, celda, bins
```

```python
folders = {"train_brain_tumor" : 1, "train_healthy": 0}
# eiquetas de las variables, donde 1 es cerebro con tumor y 0 es cerebro sano

x_hog = []
y_hog = []

for folder, label in folders.items():
    path_set = os.path.join(path_dataset, folder) # entramos a la direccion de las carpetas para entrenar
    for img_name in os.listdir(path_set): # lista de las imagenes dentro de alguna de las carpetas de entrenamiento
        path_img = os.path.join(path_set, img_name) # ingresamos el path de una de las imagenes
        img = cv2.imread(path_img) # leemos la imagen
        img = cv2.resize(img, (128,128), interpolation=cv2.INTER_AREA)
        if img is None:
            continue # si hay un archivo corrupto lo saltamos
        descriptor_hog = hog.compute(img) # creamos el descriptor
        x_hog.append(descriptor_hog) # guardamos en el vector de las x
        y_hog.append(label) #guardamos su respectiva etiqueta

np.save(os.getcwd() + "/x_train_hog", x_hog)
np.save(os.getcwd() + "/y_train_hog", y_hog)
        
```

CREAMOS AHORA CON LAS CARACTERISTICAS LBP

el set de datos train y el set de datos test

```python
x_lbp = []
y_lbp = []

for folder, label in folders.items():
    path_set = os.path.join(path_dataset, folder) # entramos a la direccion de las carpetas para entrenar
    for img_name in os.listdir(path_set): # lista de las imagenes dentro de alguna de las carpetas de entrenamiento
        path_img = os.path.join(path_set, img_name) # ingresamos el path de una de las imagenes
        img = cv2.imread(path_img) # leemos la imagen
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convertir a grises ya que asi lo necesita el LBP
        if img is None:
            continue # si hay un archivo corrupto lo saltamos
        lbp = local_binary_pattern(img, 8, 1, "uniform") # caculamos el lbp
        hist, _ = np.histogram(lbp.ravel(), 59, (0,59)) # calculamos el histograma lbp
        x_lbp.append(hist) # guardamos el histograma como descriptor en el vector de las x
        y_lbp.append(label) #guardamos su respectiva etiqueta

np.save(os.getcwd() + "/x_train_lbp", x_lbp)
np.save(os.getcwd() + "/y_train_lbp", y_lbp)
```
